# Preprocessing Data

```python
def change_gender(val):
    if 'Female' in val:
        return 0
    elif 'Male' in val:
        return 1
    else:
        return 2
        
```

```python
stroke['gender'] = stroke['gender'].apply(change_gender)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
le.fit(stroke['Residence_type'])
```

```python
stroke['Residence_type'] =  le.transform(stroke['Residence_type'])
```

```python
stroke['Residence_type'].unique()
```

```python
stroke['work_type']=le.fit_transform(stroke['work_type'])
```

```python
stroke['work_type'].unique()
```

```python
def ever_married(val):
    if 'Yes' in val:
        return 1
    else:
        return 0
```

```python
stroke['ever_married'] = stroke['ever_married'].apply(ever_married)
```

```python
stroke['smoking_status'] = le.fit_transform(stroke['smoking_status'])
```

```python
stroke.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = stroke.drop('stroke', axis=1)
y = stroke['stroke']
```

```python
X_train,X_test, y_train, y_test = train_test_split(X, y, random_state=101, test_size=0.3, shuffle=True)
```

# RandomForestClassifier

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier()
```

```python
rfc.fit(X_train, y_train)
```

```python
predict = rfc.predict(X_test)
```

```python
predict
```

```python
from sklearn import metrics
```

```python
print('r2_score', metrics.r2_score(y_test, predict))
```

```python
rfc_accuracy =  metrics.accuracy_score(y_test, predict)
rfc_accuracy
```

```python
category = ['Do not have Brain Stroke', 'Have Brain Stroke']
```

```python
own_predict = rfc.predict([[0,79.0,1,0,1,3,0,174.12,24.000000,2]])
```

```python
category[int(own_predict)]
```

# Model HyperTunning

```python
from sklearn.model_selection import GridSearchCV
```

```python
RandomForestClassifier()
```

```python
grid = {
    'n_estimators':[int(i) for i in np.linspace(start=100, stop=1200, num=6)],
    'max_depth':[int(i) for i in np.linspace(start=1, stop=120, num=6)],
    'max_features':['auto','sqrt'],
    'min_samples_split':[5,10,15,100]
}
```

```python
rfc_grid = GridSearchCV(rfc, param_grid=grid, cv=5, verbose=2, n_jobs=-1)
```

```python
rfc_grid.fit(X_train, y_train)
```

```python
rfc_grid.best_params_
```

```python
prediction=rfc_grid.predict(X_test)
```

```python
prediction
```

```python
rfc_accuracy = metrics.accuracy_score(y_test, prediction)
rfc_accuracy
```

# KNN Algorithm

```python
from sklearn.preprocessing import StandardScaler
```

```python
std = StandardScaler()
```

```python
data = std.fit_transform(X)
```

```python
df = pd.DataFrame(data, columns=X.columns)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(df, y, test_size=0.3, random_state=101, shuffle=True)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors=1)
```

```python
knn.fit(X_train, y_train)
```

```python
pre = knn.predict(X_test)
```

```python
pre
```

```python
print(metrics.accuracy_score(y_test, pre))
```

```python
print(metrics.accuracy_score(y_test, pre))
```

```python
error_rate = []

for i in range(1,50):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train, y_train)
    predict = knn.predict(X_test)
    error_rate.append(np.mean(y_test!=predict))
    
```

```python
plt.plot(range(1,50), error_rate, '-o')
```

```python
knn = KNeighborsClassifier(n_neighbors=6)
```

```python
knn.fit(X_train, y_train)
```

```python
predict = knn.predict(X_test)
```

```python
predict
```

```python
knn_accuracy = metrics.accuracy_score(y_test, predict)
knn_accuracy
```

```python
print('r2_score', metrics.r2_score(y_test, predict))
```
